```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

SUMBER DATASET : https://www.kaggle.com/datasets/shriyashjagtap/heart-attack-risk-assessment-dataset

```python
csv_path = "/content/drive/MyDrive/AVD/updated_version.csv"
df = pd.read_csv(csv_path)
df.head()
```

# ***VISUALISASI DATA***

# **1. Distribusi Usia dan Serangan Jantung**

```python
plt.figure(figsize=(8, 5))
plt.hist(df[df['heart_attack'] == 1]['age'], bins=20, color='red', edgecolor='black', alpha=0.6, label='Pernah Serangan Jantung')
plt.hist(df[df['heart_attack'] == 0]['age'], bins=20, color='blue', edgecolor='black', alpha=0.6, label='Tidak Pernah Serangan Jantung')
plt.title('Distribusi Usia dan Serangan Jantung')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

#2. **Hubungan Merokok dengan Serangan Jantung**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='smoking', hue='heart_attack', data=df, palette='pastel')
plt.title('Hubungan Merokok dengan Risiko Serangan Jantung')
plt.xlabel('Merokok (0: Tidak, 1: Ya)')
plt.ylabel('Jumlah Individu')
plt.legend(title='Serangan Jantung', labels=['Tidak', 'Ya'])
plt.show()
```

#3. **Hubungan Diabetes dengan Serangan Jantung**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='diabetes', hue='heart_attack', data=df, palette='pastel')
plt.title('Hubungan Diabetes dengan Risiko Serangan Jantung')
plt.xlabel('Diabetes (0: Tidak, 1: Ya)')
plt.ylabel('Jumlah Individu')
plt.legend(title='Serangan Jantung', labels=['Tidak', 'Ya'])
plt.show()
```

#4. **Tekanan Darah Sistolik dan Risiko Serangan Jantung**

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='heart_attack', y='systolic_bp', hue='heart_attack', data=df, palette='pastel')
plt.title('Tekanan Darah Sistolik dan Risiko Serangan Jantung')
plt.xlabel('Serangan Jantung (0: Tidak, 1: Ya)')
plt.ylabel('Tekanan Darah Sistolik')
plt.show()
```

#5. **Korelasi antara Kolesterol Total dan Serangan Jantung**

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='total_cholesterol', hue='heart_attack', kde=True, bins=20, palette='pastel', alpha=0.7)
plt.title('Distribusi Kolesterol Total Berdasarkan Serangan Jantung')
plt.xlabel('Total Kolesterol')
plt.ylabel('Frekuensi')
plt.legend(title='Serangan Jantung', labels=['Tidak', 'Ya'])
plt.show()
```

#6.**Korelasi Antar Variabel**

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df[['age', 'systolic_bp', 'diastolic_bp', 'total_cholesterol', 'ldl', 'hdl', 'smoking', 'diabetes', 'heart_attack']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```
